```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
```

```python
# I created a simple data set of haunted houses and the evidence captured
# My question is what is the most common evidence collected in haunted houses

# Import the haunted house data set
haunted = pd.read_csv('Haunted.csv') 
```

```python
haunted.head()
```

```python
# Establish the columns from my data to show (I can also resuse these in the next plots)
a = haunted['House Age']
b = haunted['Ghost Sightings']
c = haunted['Photos Captured']
d = haunted['Audio Captured']

# Create the figure
fig, ax = plt.subplots()  

# Plot my paranormal columns
ax.plot(a, label = 'House Age', 
        color = 'black')
ax.plot(b, label = 'Ghost Sightings',
       color = 'teal')  
ax.plot(c, label =' Photos Captured',
       color = 'mediumorchid')  
ax.plot(d, label = 'Audio Captured',
       color = 'darkorange')  

# X label
ax.set_xlabel('Incidents', 
             size = 15,
             color = 'mediumblue')  

# Y label
ax.set_ylabel('Count Number', 
              size = 15,
              color = 'mediumblue')  
# Title
ax.set_title('Haunted Houses', 
             size = 20, 
             color = 'darkblue')  

# Add a legend for my lines
ax.legend()  

# Show the background grid
ax.grid()
```

```python
# Sum up my data frame so that I can do a pie chart
haunt_sum = haunted.sum()

# Create new variables of the individual columns
a_sum = haunt_sum['House Age']
b_sum = haunt_sum['Ghost Sightings']
c_sum = haunt_sum['Photos Captured']
d_sum = haunt_sum['Audio Captured']
```

```python
# Plot the haunted house incidents int he pie chart
haunted_pie = [b_sum, c_sum, d_sum]
haunted_labels = 'Ghost Sightings','Photos Captured','Audio Captured'
explode = [0, .1, 0]
colors  = ['teal', 'mediumorchid', 'darkorange']

plt.pie(haunted_pie, 
        labels = haunted_labels, 
        autopct = '%1.1f%%', 
        explode = explode,
        shadow = True, 
        colors = colors)

plt.title('Haunted House Evidence', 
          size = 20, 
          color = 'midnightblue')


plt.axis('equal')

# Display the plot
plt.show()
```

```python
# Create the scatterplot using the same data as above
fig, ax = plt.subplots()
ax.scatter(b, c, d, 
           marker = 'h')
colors = ['teal', 'mediumorchid', 'darkorange']

# Plot my paranormal columns
ax.scatter(b,c,d)  
   
ax.set_title('Haunted House Evidence',
             size = 20,
             color = 'midnightblue')
# Display the background grid
ax.grid(True)
fig.tight_layout()

ax.legend(loc ='upper left', 
          title = 'Evidence')

# Show the scatterplot
plt.show()
```

#### Most Effective?
I would rank my charts in this order of effectiveness:
1. Pie Chart
